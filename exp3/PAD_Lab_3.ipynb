```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.svm import SVC
```

```python
# Creating Dataset
X,y_d = make_blobs(n_samples = 50, centers=2, random_state = 0, cluster_std=0.6) #center gives number of datasets
plt.figure(figsize=(10,10))
plt.scatter(X[:,0],X[:,1],c=y_d,s=50,cmap='autumn')
xfit = np.linspace(-1,3.5)
[plt.plot(xfit,m*xfit+b,'-k') for m,b in [(1,0.65),(0.5,1.6),(-0.2,2.9)]]
plt.xlim(-1,3.5)
```

```python
model = SVC(kernel='linear',C=1E10)
model.fit(X,y_d)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(X[:,0],X[:,1],c=y_d,s=50,cmap='autumn')
xfit = np.linspace(-1,3.5)
for m,b,d in [(1,0.65,0.33),(0.5,1.6,0.55),(-0.2,2.9,0.2)]:
  yfit = m*xfit+b
  plt.plot(xfit,yfit,'-k')
  plt.fill_between(xfit,yfit-d,yfit+d,color='pink') 
plt.xlim(-1,3.5)
```

```python
def plot_decision_function(model,ax=None,plt_support=True):
  if ax == None:
    ax = plt.gca()
    plt.rcParams["figure.figsize"] = (10,10)
  xlim = ax.get_xlim()
  ylim = ax.get_ylim()
  x = np.linspace(xlim[0],xlim[-1],30)
  y = np.linspace(ylim[0],ylim[-1],30)
  Y_mesh,X_mesh = np.meshgrid(y,X)
  xy = np.vstack([X_mesh.ravel(),Y_mesh.ravel()]).T
  P = model.decision_function(xy).reshape(X_mesh.shape)
  ax.contour(X_mesh,Y_mesh,P,colors='k',levels=[-1,0,1],alpha=0.5,linestyles=['--','-','--'])
  if plt_support:
    ax.scatter(model.support_vectors_[:,0],model.support_vectors_[:,1],s=300,facecolors='none')
  ax.set_xlim(xlim)
  ax.set_ylim(ylim)
```

```python
plt.scatter(X[:,0],X[:,1],c=y_d,s=50,cmap='autumn')
plot_decision_function(model)
```
