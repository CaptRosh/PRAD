<h3>Aim: Outlier Detection using DBSCAN</h3>

<h1>Part 1: make_blobs</h1>

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(7)
x,y = make_blobs(n_samples=200, centers=1, cluster_std=0.3)
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
dbscan = DBSCAN(eps=0.4,min_samples=20)
pred = dbscan.fit_predict(x)
```

```python
pred
```

```python
index_anomaly = np.where(pred==-1)
index_anomaly
```

```python
coords_anomaly = x[index_anomaly]
coords_anomaly
```

```python
plt.scatter(x[:,0],x[:,1])
plt.scatter(coords_anomaly[:,0],coords_anomaly[:,1],c='red')
```

<h1>Part 2: make_circles</h1>

```python
from sklearn.datasets import make_circles
from sklearn.preprocessing import StandardScaler
```

```python
x,y = make_circles(n_samples=800,factor=0.3,noise=0.4)
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
dbscan = DBSCAN(eps=0.7,min_samples=70)
pred = dbscan.fit_predict(x)
index_anomaly = np.where(pred==-1)
coords_anomaly = x[index_anomaly]
plt.scatter(x[:,0],x[:,1],c=y)
plt.scatter(coords_anomaly[:,0],coords_anomaly[:,1],c='red')
```

<h1>Part 3: make_circles</h1>

```python
from sklearn.datasets import make_moons
```

```python
x,y = make_moons(n_samples=800,noise=0.4)
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
dbscan = DBSCAN(eps=0.6,min_samples=80)
pred = dbscan.fit_predict(x)
index_anomaly = np.where(pred==-1)
coords_anomaly = x[index_anomaly]
plt.scatter(x[:,0],x[:,1],c=y)
plt.scatter(coords_anomaly[:,0],coords_anomaly[:,1],c='red')
```

<h1>Part 4: Iris Dataset</h1>

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
iris= load_iris()
pca = PCA(n_components=2)
x = pca.fit_transform(iris.data)
```

```python
dbscan = DBSCAN(eps=0.5,min_samples=5)
pred = dbscan.fit_predict(x)
index_anomaly = np.where(pred==-1)
coords_anomaly = x[index_anomaly]
plt.scatter(x[:,0],x[:,1],c=iris.target)
plt.scatter(coords_anomaly[:,0],coords_anomaly[:,1],c='red')
```
